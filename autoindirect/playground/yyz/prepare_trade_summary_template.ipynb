```python
import pandas as pd
import numpy as np
import os
```

```python
data = pd.read_csv('/d/shared/users/yyz/penfed/data_3rd_party_refi_f_sent_withunfundedtarget_zestscore_vs4_20250618.csv')
```

```python
data['appmonth'] = data['Application_Date'].str[:7]
data['bookmonth'] = data['Funded_Date'].str[:7]
```

```python
funded = data[data['Decision'] == 'Funded']
```

```python
funded_summary = (
    funded.groupby('appmonth')
      .agg(
        total_count       = ('appmonth','size'),
        mob_6_60_notna    = ('mob_6_60',lambda x: x.notna().sum()),
        mob_6_60_eq_1     = ('mob_6_60',lambda x: (x == 1).sum()),
        mob_12_90_notna    = ('mob_12_90',lambda x: x.notna().sum()),
        mob_12_90_eq_1     = ('mob_12_90',lambda x: (x == 1).sum())
      )
      .reset_index()
)

display(funded_summary)
```

```python
funded_summary = (
    funded.groupby('bookmonth')
      .agg(
        total_count       = ('bookmonth','size'),
        mob_6_60_notna    = ('mob_6_60',lambda x: x.notna().sum()),
        mob_6_60_eq_1     = ('mob_6_60',lambda x: (x == 1).sum()),
        mob_12_90_notna    = ('mob_12_90',lambda x: x.notna().sum()),
        mob_12_90_eq_1     = ('mob_12_90',lambda x: (x == 1).sum())
      )
      .reset_index()
)

display(funded_summary)
```
